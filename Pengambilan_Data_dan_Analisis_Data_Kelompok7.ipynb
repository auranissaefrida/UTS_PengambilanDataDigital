# **Data Preprocessing Jam Flash Sale Jam 06.00-10.00**

```python
import requests
import json
import pandas as pd


url = ("https://raw.githubusercontent.com/auranissaefrida/UTS_PengambilanDataDigital/main/response_0930.txt")
            

# Make a GET request to the URL and retrieve the content as text
response = requests.get(url)
data_0930 = response.text
```

```python
json_data_0930 = json.loads(response.text)
data_0930 = pd.json_normalize(json_data_0930[3]['data']['componentInfo']['data']['component']['data'])
data_0930
```

```python
df_0930=data_0930[['name','status','campaign_code','campaign_sold_count','stock_sold_percentage','custom_stock','discounted_price','discount_percentage', 'price','shop_name','shop_location']]
df_0930
```

```python
df_0930['date'] = df_0930['campaign_code'].str[3:14]
df_0930.drop('campaign_code', axis=1, inplace=True)
df_0930.sample(10)
```

```python
print(df_0930.dtypes)
```

```python
# Define a function to convert a price string to an integer
def price_to_int(price_str):
    # Remove the "Rp " prefix and the "." separator
    price_str = price_str.replace("Rp ", "").replace(".", "")
    # Convert the resulting string to an integer
    return int(price_str)

# Apply the function to the "discounted_price" column
df_0930['discounted_price'] = df_0930['discounted_price'].apply(price_to_int)
df_0930['price'] = df_0930['price'].apply(price_to_int)
```

```python
print(df_0930.dtypes)
```

```python
df_0930.isnull().all()
```

```python
df_0930.to_csv('data_0930.csv', index=False)
```

# Mengubah data frame menjadi SQL

```python
import pandas as pd

def dtype_map(dtype):
    if 'int' in str(dtype):
        return 'INTEGER'
    elif 'float' in str(dtype):
        return 'FLOAT'
    elif 'datetime' in str(dtype):
        return 'DATETIME'
    else:
        return 'VARCHAR(255)'

columns = ', '.join([f'{col} {dtype_map(dtype)}' for col, dtype in df.dtypes.items()])
create_table = f'CREATE TABLE my_table ({columns});\n'

# generate the INSERT INTO statements
insert_into = ''
for index, row in df.iterrows():
    values = ', '.join([f"'{val}'" if isinstance(val, str) else str(val) for val in row])
    insert_into += f"INSERT INTO my_table ({', '.join(df.columns)}) VALUES ({values});\n"

# save the SQL statements to a file
with open('my_table.sql', 'w') as f:
    f.write(create_table)
    f.write(insert_into)
```
